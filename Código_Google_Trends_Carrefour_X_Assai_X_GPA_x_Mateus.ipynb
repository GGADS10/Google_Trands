```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pytrends
from pytrends.request import TrendReq

# Cria o objeto TrendReq
pytrends = TrendReq()

# Define as palavras-chave e o idioma
keywords = ["Carrefour", "Assai Atacadista", "Grupo Mateus", "Pão de Açúcar"]
pytrends.build_payload(keywords, cat=0, timeframe='today 5-y', geo='BR', gprop='')

# Obtém os dados das tendências de busca
data = pytrends.interest_over_time()

# Filtra os dados para o período desejado
data = data.loc['2018-01-01':]

# Plota as linhas separadas para cada busca
plt.plot(data.index, data['Carrefour'], label='Carrefour')
plt.plot(data.index, data['Assai Atacadista'], label='Assai Atacadista')
plt.plot(data.index, data['Grupo Mateus'], label='Grupo Mateus')
plt.plot(data.index, data['Pão de Açúcar'], label='Pão de Açúcar')

# Formatação das datas
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())

# Rotação dos rótulos das datas
plt.gcf().autofmt_xdate()

# Exibe a legenda e o gráfico
plt.legend()
plt.show()


```
